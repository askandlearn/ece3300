```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0,2*np.pi, 100)
v = 4*np.cos(3*t)
i = 1/12 * np.sin(3*t)
p = v*i
```

```python
fig, ax = plt.subplots(3,1, sharex=True, dpi=150)
ax[0].plot(t,v)
ax[0].set_ylabel('Voltage (V)')
ax[0].grid('on')
ax[1].plot(t,i)
ax[1].set_ylabel('Current (A)')
ax[1].grid('on')
ax[2].plot(t,p)
ax[2].set_ylabel('Power (W)')
ax[2].grid('on')
ax[2].set_xlabel('Time (s)')
```

```python
fig, ax0 = plt.subplots(dpi=150)
ax1 = ax0.twinx()
ax0.plot(t,v)
ax0.set_ylabel('Voltage (V)')
ax1.plot(t,i, 'r')
ax0.legend(['Voltage'])
ax1.set_ylabel('Current (A)')
ax1.set_xlabel('Time (s)')
ax1.legend(['Current'])
```
